```python
from sklearn.ensemble import RandomForestClassifier
```

```python
train=pd.get_dummies(train)
test=pd.get_dummies(test)
```

```python
train.shape, test.shape
```

```python
rf=RandomForestClassifier(random_state=42,n_estimators=500)
rf.fit(train,target)
```

```python
pred=rf.predict(test)
```

```python
# 교차검증 성능확인용
from sklearn.model_selection import cross_val_score
```

```python
rf2=RandomForestClassifier(random_state=42,n_estimators=500)
```

```python
# help(cross_val_score)
# array([0.48837209, 0.47861965, 0.47861965, 0.47711928, 0.49587397])
score= cross_val_score(rf2,train,target,cv=5,scoring='f1_macro')
score
```
